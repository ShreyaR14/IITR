```python
# Read image - fishplate 1
fp1 = cv2.imread('C:/Users/hp/Desktop/Mod1-IITR/fishplate/pos/3.jpg') 
#RGB 
plt.imshow(fp1)
```

```python

# convert the image to HSV and create a histogram
rgb = cv2.cvtColor(fp1, cv2.COLOR_BGR2RGB)  
plt.imshow(rgb)
hist = cv2.calcHist([rgb], [0], None, [256], [0, 256])
plt.figure()
plt.title("RGB Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.xlim([0, 256])

for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
```
